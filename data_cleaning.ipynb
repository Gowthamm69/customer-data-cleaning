```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# ----------------------------
# STEP 0: CREATE DATA
# ----------------------------
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

print("Original Data:\n", df)

# ----------------------------
# TASK 1: CLEAN DATA
# ----------------------------

# Fill missing age with median
df['age'] = df['age'].fillna(df['age'].median())

# Fill missing city with mode
df['city'] = df['city'].fillna(df['city'].mode()[0])

# Remove duplicate rows
df = df.drop_duplicates()

print("\nAfter Cleaning:\n", df)

# ----------------------------
# TASK 2: ENCODE DATA
# ----------------------------

# One-Hot Encoding for city
df = pd.get_dummies(df, columns=['city'])

# Label Encoding for gender
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

print("\nAfter Encoding:\n", df)

# ----------------------------
# TASK 3: SCALING
# ----------------------------

# Columns to scale
cols = ['age', 'annual_income']

# Min-Max Scaling
df_minmax = df.copy()
mm = MinMaxScaler()
df_minmax[cols] = mm.fit_transform(df_minmax[cols])

print("\nMin-Max Scaled Data:\n", df_minmax)

# Standardization
df_standard = df.copy()
ss = StandardScaler()
df_standard[cols] = ss.fit_transform(df_standard[cols])

print("\nStandard Scaled Data:\n", df_standard)
```
